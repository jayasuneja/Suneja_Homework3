```python
# 1. Import the Data

# mounting drive
from google.colab import drive
drive.mount('/content/drive')
filedir = '/content/drive/MyDrive/FWE458/'
drive.mount("/content/drive", force_remount=True)

# importing data
import pandas as pd
fname  = filedir + "MaunaLoaCO2_monthly_sitename.csv"
DF = pd.read_csv(fname)

# printing head of data
print(DF.head())
```

```python
# 2. Filter Data

# Exclude data from the years 1958 and 2022
DF = DF[~DF['Year'].isin([1958,2022])]
  # This line of code finds the 'Year' column in dataframe DF (DF['Year'])
  # Then it finds if 1958 and 2022 values exist in that column (.isin)
  # Then it reverses the action where true becomes false and it excludes those years (~)
print(DF)

# Remove rows where "numofdays" or "stdofdays" contain invalid (negative) values
DF = DF.dropna().loc[(DF["numofdays"] >= 0) & (DF["stdofdays"] >= 0)]
  # This line of code drops any rows containing NaN values (.dropna)
  # Then it finds the 'numofdays' column and drops the negative values (>=0)
  # Then it finds the 'stdofdays' column and drops the negative values (>=0)
  # The & indicates that it will drop rows if either column has negative values
print(DF.head())

```

```python
# 3. Compute Annual Statistics
DF['Date'] = pd.to_datetime(DF[['Year', 'Month']].assign(day=1))
DF.drop(columns="site", inplace=True,errors="ignore")
DF.drop(columns="Decimal_date", inplace=True, errors = "ignore")
DF = DF[['Date','Year','Month','Monthly_average','deseasonalized','numofdays','stdofdays']]
DF
# Calculate the annual average CO2 concentration from "Monthly_average"
# and store it in "Annual_average"
DF["Annual_average"] = DF.groupby("Year")["Monthly_average"].transform("mean")
  # This makes a new column called "Annual_average"
  # Then it groups all the values in "Monthly_average" by "Year"
  # Finally, it takes the mean of those groups and stores it in the new column

# Calculate the annual average deseasonalized CO2 concentration from "deseasonalized"
# and store it in "Deseasonalized_annual_average"
DF["Deseasonalized_annual_average"] = DF.groupby("Year")["deseasonalized"].transform("mean")
  # This line of code is similar to the last
  # But it's grouping the values in "deseasonalized" by year and averaging them

# Compute the standard deviation of the 12-month "Monthly_average" values for each year
# and store it in "std_of_months"
DF["std_of_months"] = DF.groupby("Year")["Monthly_average"].transform("std")
  # Similar to the previous two lines of code, this groups the monthly averages
  # Then takes the standard deviation of those groupings

print(DF)
```

```python
# 4. Apply Custom Function

# Write a function to classify CO2 levels into categories:
  # "Low" if CO2 is below 350 ppm
  # "Moderate" if CO2 is between 350 and 400 ppm
  # "High" if CO2 is above 400 ppm
# Apply this function to create a new column "CO2_Category"

def categorize_co2(value):
  if value < 350:
    return 'Low'  # if the monthly aveg is below 350ppm, category is low
  elif value < 400:
    return 'Moderate' # otherwise, if the monthly ppm is 350<x<400, category is moderate
  else:
    return 'High' # otherwise, if the monthly ppm is >400, category is high

# Applying the function to the 'Monthly_average' column and
# making a new column to store the results in
DF['CO2_Category'] = DF['Monthly_average'].apply(categorize_co2)

DF

```
